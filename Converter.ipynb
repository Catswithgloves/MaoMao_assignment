## Create converting function to be called

```python
# temperature function
def ftoc(fahrenheit):
    celsius=(fahrenheit-32)*5/9
    return celsius

def ctof(celsius):
    fahrenheit=(celsius*9/5)+32
    return fahrenheit

#length function
def mtokm(m):
    km=m*1.60934
    return km

def kmtom(km):
    m=km/0.6213711922
    return m

#weight function
def ptokilo(p):
    kilo=p*0.45359237
    return kilo

def kilotop(kilo):
    p=kilo*2.2046226218488
    return p

#Ask user if they want to start again, The startswith() method returns True if user input starts with Y,otherwise False
def startagain():
    print('Do you want to start again? (yes or no)')
    return input().lower().startswith('y')
```
